```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'genres': ['Action', 'Comedy', 'Action', 'Drama', 'Comedy', 'Drama', 'Action']
})

genre_count = df['genres'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=genre_count.index, y=genre_count.values, palette='viridis')
plt.title('Janr üzrə film sayı')
plt.xlabel('Janr')
plt.ylabel('Film sayı')
plt.show()
```

```python
df['release_year'] = [2010, 2011, 2010, 2012, 2011, 2010, 2012]

year_count = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.lineplot(x=year_count.index, y=year_count.values, marker='o')
plt.title('İllərə görə film sayı-Movie counts by year')
plt.xlabel('İl-Year')
plt.ylabel('Film sayı-Movie count')
plt.show()
```

```python
df['budget'] = [1000000, 500000, 1500000, 2000000, 800000, 1200000, 2500000]
df['revenue'] = [5000000, 1000000, 7000000, 8000000, 2000000, 6000000, 9000000]

plt.figure(figsize=(8,5))
sns.scatterplot(x='budget', y='revenue', data=df, hue='genres', s=100)
plt.title('Büdcə vs Gəlir')
plt.xlabel('Büdcə')
plt.ylabel('Gəlir')
plt.show()
```
